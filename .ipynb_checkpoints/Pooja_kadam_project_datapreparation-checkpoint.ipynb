```python
def box_plotter(df, features, target):
    
    plt.figure(figsize=(15, 30))
    for (i, feature) in zip(range(len(features)), features):
        plt.subplot(int('{}2{}'.format(len(features), i+1)))
        sns.boxplot(x=target, y=feature, data=df,
                    width=0.5, flierprops={'marker': 'x'})
        plt.title(label='Boxplot of {}'.format(feature), fontsize=10)
        plt.xlabel(xlabel=None)
        plt.ylabel(ylabel=None)
        i += 1
    plt.show()
```

```python
imp_cols = ['alpha', 'delta', 'u', 'g', 'r', 'i', 'z', 'redshift']
display(star_data[imp_cols].describe().T)
```

```python
box_plotter(df=star_data, features=imp_cols, target='class')
```

```python
sns.countplot(x = star_data["class"], palette="Set3")
plt.title("Class Distribution",fontsize=10)
plt.show()
```

```python
star_pairplot = sns.pairplot(data=star_data[['alpha', 'delta', 'redshift','u', 'g', 'r', 'i', 'z', 'class']], hue='class')
star_pairplot.fig.suptitle(t='Pair Plot - Stellar Data', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.heatmap(data=star_data[['alpha', 'delta', 'redshift','u', 'g', 'r', 'i', 'z']].corr(),
            annot=True, fmt='.2f', linewidths=0.1)
plt.show()
```
